```python
import pandas          as pd
import numpy           as np
import sklearn.metrics as mt

from sklearn.tree          import DecisionTreeRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, MinMaxScaler
from sklearn.linear_model  import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
path_xtrain = 'datasets//X_training.csv'
path_ytrain = 'datasets//y_training.csv'

X_train = pd.read_csv( path_xtrain )
y_train = pd.read_csv( path_ytrain )

path_xtest = 'datasets//X_test.csv'
path_ytest = 'datasets//y_test.csv'

X_test = pd.read_csv( path_xtest )
y_test = pd.read_csv( path_ytest )

path_xval = 'datasets//X_validation.csv'
path_yval = 'datasets//y_val.csv'

X_val = pd.read_csv( path_xval )
y_val = pd.read_csv( path_yval )
```

```python
features = ['song_duration_ms', 'acousticness', 'danceability', 'energy',
    'instrumentalness', 'key', 'liveness', 'loudness',
    'speechiness', 'tempo', 'audio_valence']

X_train = X_train.loc[:, features]
X_test = X_test.loc[:, features]
X_val = X_val.loc[:, features]
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.transform(X_val)
```

# Polinomial Elastic Net

from sklearn.linear_model import ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import numpy as np

degree = np.arange(2, 5, 1)
alpha = [0.005, 0.009, 0.01]
l1_ratio = [0.01, 0.1, 0.5, 1]
max_iter = [27, 28, 30]

melhor_degree = None
melhor_alpha = None
melhor_iter = None
melhor_l1_ratio = None
melhor_desempenho = float('inf')


model = ElasticNet()

for d in degree:
    # Criar o modelo de features polinomiais
    poly_model = PolynomialFeatures(degree=d, include_bias=False)

    # Ajustar e transformar os dados de treinamento
    X_poly_train = poly_model.fit_transform(X_train)

    # Transformar os dados de validação
    X_poly_val = poly_model.transform(X_val)

    for a in alpha:
        for m in max_iter:
            for l in l1_ratio:
                # Configurar os hiperparâmetros do modelo
                model.set_params(alpha=a, l1_ratio=l, max_iter=m)

                # Ajustar o modelo aos dados de treinamento
                model.fit(X_poly_train, y_train)

                # Fazer previsões nos dados de validação
                y_pred = model.predict(X_poly_val)

                # Calcular métricas de avaliação
                mse = mt.mean_squared_error(y_val, y_pred)
                rmse = mt.root_mean_squared_error(y_val, y_pred)
                mae = mt.mean_absolute_error(y_val, y_pred)
                mape = mt.mean_absolute_percentage_error(y_val, y_pred)

                # Calcular o desempenho médio
                desempenho_atual = (mse + rmse + mae + mape) / 4

                # Atualizar o melhor desempenho e os melhores hiperparâmetros
                if desempenho_atual < melhor_desempenho:
                    melhor_desempenho = desempenho_atual
                    melhor_degree = d
                    melhor_alpha = a
                    melhor_iter = m
                    melhor_l1_ratio = l

                    print('Média de erros:', desempenho_atual)
                    print('Degree:', d)
                    print('Alpha:', a)
                    print('Max_iter:', m)
                    print('L1_ratio:', l)

print("Melhor valor de degree:", melhor_degree)
print("Melhor valor de alpha:", melhor_alpha)
print("Melhor valor de max_iter:", melhor_iter)
print("Melhor valor de l1_ratio:", melhor_l1_ratio)
print("Melhor desempenho (Média das métricas):", melhor_desempenho)
